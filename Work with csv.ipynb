# Read TSV File

```python
pd.read_csv('demo_mlexpec.tsv')
```

# Separate Parameter

```python
pd.read_csv('demo_mlexpec.tsv', sep='\t')
```

# Columns Parameters

```python
pd.read_csv('demo_mlexpec.tsv', sep='\t', names =['Unit', 'Sex', 'Age', 'Time', 'Years'])
```

# Indx_Col Parameter

```python
 pd.read_csv('Screen_Time.csv', index_col = 'Age')
```

```python
pd.read_csv('customers-1000.csv', index_col = 'Index')
```

# Header Paramaeter

```python
 pd.read_csv('customers-1000.csv', header =1)
```

# Header parameter

```python
pd.read_csv('customers-1000.csv', header= 1)
```

 # Use_cols Parameter

```python
pd.read_csv('customers-1000.csv', usecols = ['Customer Id', 'First Name', 'Company', 'Country']) 
```

# skiprows / nrows Parameter

```python
pd.read_csv('customers-1000.csv', skiprows = [0,1])
```

```python
pd.read_csv('customers-1000.csv', nrows = 100)
```

# Encoding Parameter

```python
pd.read_csv('zomato.csv', encoding = 'latin-1')
```

# Skip bad lines

```python
pd.read_csv('zomato.csv', sep=';', encoding='latin-1', error_bad_lines = False)
```

# dtypes

```python
pd.read_csv('customers-1000.csv', dtype={'Index':float})
```

```python
pd.read_csv('customers-1000.csv', dtype={'Index':float}).info()
```

# Date Handling

```python
pd.read_csv('customers-1000.csv').info()
```

```python
pd.read_csv('customers-1000.csv', parse_dates=['Subscription Date'])
```

```python
pd.read_csv('customers-1000.csv', parse_dates=['Subscription Date']).info()
```
